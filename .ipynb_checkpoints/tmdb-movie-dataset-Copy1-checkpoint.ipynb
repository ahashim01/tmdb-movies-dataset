### Data Cleaning:
> <ol>
    <li>we will drop the unused columns from our data set ['id', 'imdb_id', 'homepage', 'tagline', 'keywords', 'overview']</li>
    <li>check duplicated rows and drop duplicated</li>
    <li>Drop NaN values</li>
    <li>we need to fill the zero values with the mean of each column</li>
</ol>

###### 1.we will drop the unused columns from our data set ['id', 'imdb_id', 'homepage', 'tagline', 'keywords', 'overview']

```python
df.drop(columns=['id', 'imdb_id', 'homepage', 'tagline', 'keywords', 'overview'], inplace=True)
df.info()
```

###### 2.check duplicated rows and drop duplicated

```python
# check duplicated
df.duplicated().sum()
```

```python
# Drop duplicate
df.drop_duplicates(inplace=True)
```

###### 3.Drop NaN values

```python
df.dropna(inplace=True)
```

###### 4.we need to fill the zero values with the mean of each column

```python
# Fill zeros in each column with the mean value of the column
df.popularity = df.popularity.mask(df.popularity == 0, df.popularity.mean())
df.budget = df.budget.mask(df.budget == 0, df.budget.mean())
df.revenue = df.revenue.mask(df.revenue == 0, df.revenue.mean())
df.vote_count = df.vote_count.mask(df.vote_count == 0, df.vote_count.mean())
df.budget_adj = df.budget_adj.mask(df.budget_adj == 0, df.budget_adj.mean())
df.revenue_adj = df.revenue_adj.mask(df.revenue_adj == 0, df.revenue_adj.mean())
```

```python
df.info()
```

## Exploratory Data Analysis

### Research Question 1: Which geners are produced from year to year ?

```python
# function to get unique genres
def get_data(data, column):
    data = data[column].str.cat(sep='|')
    data = pd.Series(data.split('|')).str.strip()
    return data
```

```python
# get the unique genres
df.genres = df.genres.apply(str)
genres = df.genres.str.cat(sep="|")
unique_genres = pd.Series(data=genres.split("|")).str.strip()
unique_genres.value_counts()
```

###### Show the rate of production of each genre over year

```python
for genre in unique_genres.unique():
    # create a dataframe for each genre
    genre_df = df[df.genres.str.contains('{}'.format(genre))]
    genre_df.groupby('release_year')['genres'].count().plot(kind='bar', figsize=(15,10))
    plt.xlabel('Year', fontsize=10)
    plt.ylabel('Movies Count', fontsize=10)
    plt.title('{} Movies over Year'.format(genre))
    plt.show();
```

### Research Question 2: Which are the highest profitable, popular movie genre?

###### we will get the profit from the '_adj'  of the budget and revenue columns as they show in terms of 2010 dollors, accounting for inflation over time

```python
df['profit'] = df.revenue_adj - df.budget_adj
df.info()
```

```python
# Create profit sum for each genre
genres = list(unique_genres.unique())
genres_profit = []
for genre in unique_genres.unique():
    genre_profit = df[df.genres.str.contains('{}'.format(genre))]
    genres_profit.append(genre_profit.profit.sum())
genres_profit_dict = {'genres': genres, 'profit': genres_profit}
profit_data = pd.DataFrame(genres_profit_dict)
profit_data.sort_values('profit', inplace=True)
profit_data.set_index('genres', inplace=True)
profit_data
```

```python
profit_data.plot(kind='bar', figsize=(10,8), color='green')
plt.xlabel('Movie Genres', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.title('Total Profits of the geners over years');
```

```python
# Get the genres with the high popularity
highest_popularity = df[df.popularity > df.popularity.mean()]
highest_popularity_genres = highest_popularity.genres.str.cat(sep="|")
highest_popularity_unique_genres = pd.Series(data=highest_popularity_genres.split("|")).str.strip()
highest_popularity_unique_genres.value_counts()
```

```python
highest_popularity_unique_genres.value_counts(ascending=True).plot(kind='bar', figsize=(10,8), color='indigo')
plt.xlabel('Movie Genres', fontsize=12)
plt.ylabel('No. of Movies', fontsize=12)
plt.title('Movies Genres with Highest Popularity');
```
